```python
def get_player_names():
    player_1 = None
    while not player_1:
        player_1=input("Enter the name of player 1: ").strip()
        
    player_2= None
    while not player_2:
        player_2=input("Enter the name of player 2: ").strip()
        
    print("\nMatch will be played between '" + player_1 + "' and '" + player_2 + "'\n")
    return player_1, player_2
```

```python
def get_dice_value():
    
    dice_value = random.randint(1, DICE_FACE)
    print("You got " + str(dice_value))
    return dice_value

    
def snake_ladder_game(player_name, current_value, dice_value):
    
    old_value = current_value
    current_value = current_value + dice_value
    
    if current_value > max_value:
        print("You need " + str(max_value - old_value) + " to win this game. Keep trying...")
        return old_value
    
    print( player_name + " moved from " + str(old_value) + " to " + str(current_value))
    if current_value in Position_of_snakes:
        final_value = Position_of_snakes.get(current_value)
        print("Shit! You got a snake bite")
        print( player_name + " Down from " + str(current_value) + " to " + str(final_value))
        
    elif current_value in Position_of_ladders :
        final_value = Position_of_ladders.get(current_value)
        print("Yayyy! You got a ladder ")
        print( player_name + " climbed the ladder from " + str(current_value) + " to " + str(final_value))

    else:
        final_value = current_value

    return final_value
```

```python
def check_win_the_player(player_name,position):
    if max_value==position:
        print("\n\n\nThats it. \n\n" + player_name + " won the game.")
        print("Congratulations " + player_name)
        print("\nThank you for playing the game.")
        return True
```
